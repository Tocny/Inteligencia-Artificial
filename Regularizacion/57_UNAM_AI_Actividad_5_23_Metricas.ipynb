# Métricas de clasificación y regresión (con fórmulas)
  
>
> **Métricas para Clasificación (Matriz de Confusión):**  
> - *Exactitud (Accuracy)*: porcentaje total de predicciones correctas.  
>   $$ \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN} $$  
> - *Precisión*: de los casos predichos como positivos, ¿cuántos lo eran realmente?  
>   $$ \text{Precision} = \frac{TP}{TP + FP} $$  
> - *Recall*: de los positivos reales, ¿cuántos fueron detectados?  
>   $$ \text{Recall} = \frac{TP}{TP + FN} $$  
> - *F1-Score*: media armónica entre precisión y recall.  
>   $$ F_1 = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}} $$  
>
> **Métricas para Regresión (basadas en el error):**  
> - *MSE (Error Cuadrático Medio)*: penaliza errores grandes.  
>   $$ \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$  
> - *RMSE*: raíz del MSE, en las mismas unidades que $y$.  
>   $$ \text{RMSE} = \sqrt{\text{MSE}} $$  
> - *MAE (Error Absoluto Medio)*: menos sensible a outliers.  
>   $$ \text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i| $$  
> - *$R^2$ (Coeficiente de Determinación)*: proporción de varianza explicada.  
>   $$ R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2} $$
