# **Razonamiento Probabilístico y Modelos Ocultos de Markov**


## Superioridad del razonamiento probabilístico sobre la lógica de primer orden   
> La lógica de primer orden solo permite afirmar que algo es verdadero o falso. Sin embargo, en el mundo real la información suele ser incompleta, incierta o tener ruido. El razonamiento probabilístico permite asignar grados de creencia a los eventos y tomar decisiones racionales aun cuando no se tiene certeza absoluta.

## Diferencia entre incertidumbre aleatoria e incertidumbre epistémica

> La incertidumbre aleatoria proviene de la variabilidad natural del fenómeno y existe aunque conozcamos perfectamente el sistema. La incertidumbre epistémica aparece por falta de información, conocimiento o datos suficientes, por lo que puede reducirse obteniendo más evidencia.

## Proceso de “Marginalización”

> La marginalización consiste en eliminar una variable que no nos interesa sumando todas sus posibles probabilidades. De esta forma obtenemos la probabilidad de una variable específica sin considerar los demás factores del problema.

## Definición de Experimento Aleatorio

> Un experimento aleatorio es cualquier proceso cuyo resultado no puede conocerse con certeza antes de realizarse. Algunos ejemplos son los juegos de azar, procesos económicos o estudios médicos.

## Diferencia entre modelos discriminativos y generativos

> Los modelos discriminativos aprenden directamente la probabilidad de una clase dada la información observada, es decir, $P(Y|X)$. En cambio, los modelos generativos modelan primero la distribución conjunta $P(X,Y)$ y posteriormente utilizan el Teorema de Bayes para obtener la probabilidad condicional.

## Nodos y arcos en una Red Bayesiana

> En una Red Bayesiana, los nodos representan variables aleatorias y los arcos o flechas representan relaciones de dependencia o influencia causal directa entre esas variables. Esto permite modelar cómo un evento puede afectar a otro.

## Eficiencia computacional de Naive Bayes

> Naive Bayes es eficiente porque solo necesita recorrer una vez los datos para calcular estadísticas básicas como medias, desviaciones y probabilidades previas. Además, asume independencia entre las características, lo que simplifica mucho los cálculos y reduce el costo computacional.

## “Verosimilitud” en Naive Bayes Gaussiano

> La verosimilitud mide qué tan probable es observar un dato dentro de una clase determinada. En Naive Bayes Gaussiano se calcula mediante una distribución normal; mientras más cerca esté el valor del promedio de la clase, mayor será su verosimilitud.

## Condición de “Irreducible” en una Cadena de Markov

> Una Cadena de Markov es irreducible cuando es posible llegar de cualquier estado a cualquier otro en un número finito de pasos. En otras palabras, todos los estados están conectados y ninguno queda aislado.

## Propósito del algoritmo de Viterbi en Modelos Ocultos de Markov

> El algoritmo de Viterbi sirve para encontrar la secuencia de estados ocultos más probable que generó una secuencia de observaciones. Su objetivo es identificar la ruta óptima dentro de un Modelo Oculto de Markov.

#  **Aprendizaje Automático, Métricas, K-NN, SVM, Árboles, Redes Neuronales, Transformers**

## Definición de Aprendizaje Automático vs programación tradicional

> En la programación tradicional el programador define explícitamente las reglas que debe seguir el sistema para resolver un problema. En el Aprendizaje Automático, en lugar de programar cada regla, se utiliza un algoritmo que aprende patrones a partir de los datos y construye un modelo capaz de realizar predicciones o tomar decisiones.

## Diferencia entre paradigma conexionista y simbólico

> El paradigma simbólico representa el conocimiento mediante reglas lógicas, símbolos y razonamientos que pueden explicarse fácilmente. En cambio, el paradigma conexionista representa el conocimiento mediante conexiones y pesos numéricos, logrando gran capacidad de aprendizaje, aunque muchas veces funciona como una “caja negra” difícil de interpretar.

## Aprendizaje automático para tareas de percepción sensorial

> **¿Por qué se dice que el aprendizaje automático es mejor para tareas de percepción sensorial?**

> Porque tareas como reconocer imágenes, voz o sonidos son procesos que los humanos realizamos de forma casi automática y que resultan difíciles de describir mediante reglas precisas. El Aprendizaje Automático puede aprender estos patrones a partir de grandes cantidades de ejemplos, obteniendo mejores resultados que una programación basada únicamente en reglas.

## Concepto de “Generalización” mediante la analogía del examen

> La generalización es la capacidad de un modelo para responder correctamente preguntas nuevas. Es similar a un estudiante que comprende un tema y puede resolver preguntas nuevas en un examen. Si solo memoriza respuestas específicas, tendrá dificultades cuando aparezcan problemas diferentes.

## Diferencia entre aprendizaje supervisado y no supervisado

> El aprendizaje supervisado utiliza datos etiquetados, es decir, ejemplos donde ya se conoce la respuesta correcta, para aprender a realizar predicciones. Por otro lado, el aprendizaje no supervisado trabaja con datos sin etiquetas y busca descubrir patrones, relaciones o agrupaciones ocultas dentro de la información disponible.

## Ciclo de interacción en el Aprendizaje por Refuerzo

> En el Aprendizaje por Refuerzo, un agente observa el estado del entorno, selecciona una acción y recibe una recompensa o penalización según el resultado obtenido. Después pasa a un nuevo estado y repite el proceso, aprendiendo gradualmente qué acciones le permiten maximizar la recompensa acumulada a largo plazo.

## Importancia de la aleatoriedad al obtener instancias para test

> La aleatoriedad ayuda a que el conjunto de entrenamiento y el de prueba sean representativos de la misma población. Esto evita sesgos o diferencias sistemáticas entre ambos grupos y permite evaluar de manera más justa la capacidad de generalización del modelo.

## Definición de “fold” en validación cruzada

> Un fold es una de las particiones en las que se divide el conjunto de datos durante la validación cruzada. En cada iteración, un fold se utiliza para probar el modelo y los demás para entrenarlo, permitiendo evaluar su desempeño de forma más confiable y aprovechando todos los datos disponibles.

## Métricas de clasificación y regresión (con fórmulas)
  
>
> **Métricas para Clasificación (Matriz de Confusión):**  
> - *Exactitud (Accuracy)*: porcentaje total de predicciones correctas.  
>   $$ \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN} $$  
> - *Precisión*: de los casos predichos como positivos, ¿cuántos lo eran realmente?  
>   $$ \text{Precision} = \frac{TP}{TP + FP} $$  
> - *Recall*: de los positivos reales, ¿cuántos fueron detectados?  
>   $$ \text{Recall} = \frac{TP}{TP + FN} $$  
> - *F1-Score*: media armónica entre precisión y recall.  
>   $$ F_1 = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}} $$  
>
> **Métricas para Regresión (basadas en el error):**  
> - *MSE (Error Cuadrático Medio)*: penaliza errores grandes.  
>   $$ \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$  
> - *RMSE*: raíz del MSE, en las mismas unidades que $y$.  
>   $$ \text{RMSE} = \sqrt{\text{MSE}} $$  
> - *MAE (Error Absoluto Medio)*: menos sensible a outliers.  
>   $$ \text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i| $$  
> - *$R^2$ (Coeficiente de Determinación)*: proporción de varianza explicada.  
>   $$ R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2} $$

## Decisión de etiqueta en k-NN

> k-NN calcula la distancia entre el nuevo punto y todos los puntos del conjunto de entrenamiento. Después selecciona los $k$ vecinos más cercanos, observa las etiquetas de esos vecinos y asigna la clase que tenga la mayoría de votos. Por eso se considera un algoritmo basado en similitud o cercanía entre observaciones.

## Kernel Lineal vs Kernel Polinomial

> Un Kernel Lineal intenta separar los datos mediante una línea recta o un hiperplano. En cambio, un Kernel Polinomial permite generar fronteras de decisión curvas o más complejas, por lo que puede manejar mejor problemas donde los datos no son linealmente separables.

## Principal desventaja de SVM con datasets grandes

> La principal desventaja es que el entrenamiento y el uso de memoria aumentan considerablemente conforme crece el tamaño del dataset. Por ello, las SVM pueden volverse lentas e ineficientes cuando se trabaja con millones de registros.

## Diferencia entre nodo raíz, interno y hoja en un árbol de decisión

> El nodo raíz contiene todo el conjunto de datos y representa la primera decisión del árbol. Los nodos internos realizan divisiones adicionales evaluando atributos específicos. Las hojas son los nodos finales y contienen la predicción o decisión que emite el modelo.

## Aprendizaje en una red neuronal mediante ajuste de pesos

> La red neuronal realiza una predicción, la compara con el resultado correcto y calcula el error cometido. Después ajusta los pesos de sus conexiones para reducir ese error, mejorando gradualmente su desempeño en futuras predicciones.

## Problema que resuelven los Transformers frente a redes recurrentes

> **¿Qué problema resuelven los Transformers en comparación con las redes recurrentes tradicionales?**  
> Los Transformers solucionan las limitaciones de memoria y del procesamiento secuencial de las redes recurrentes. Gracias al mecanismo de atención, pueden analizar todo el texto al mismo tiempo y capturar relaciones entre palabras que están muy alejadas entre sí.

# **Ética, Procesamiento de Lenguaje Natural, Visión Computacional, AGI, Multimodalidad y ODS**


## Interpretabilidad vs Explicabilidad

> La interpretabilidad es la capacidad de comprender directamente cómo las variables de entrada influyen en la salida de un modelo, permitiendo entender la relación causa-efecto. La explicabilidad, en cambio, consiste en justificar o comunicar por qué el modelo produjo un resultado específico, aun cuando su funcionamiento interno sea complejo o difícil de entender.

## Tres fases del proceso de percepción en sistemas artificiales

> Las 3 etapas son:  
> 1. *Captura*: obtención de señales del entorno mediante sensores como cámaras, micrófonos o sensores de temperatura.  
> 2. *Representación*: conversión de esas señales en estructuras digitales como vectores, matrices o embeddings.  
> 3. *Interpretación*: asignación de significado a los datos mediante clasificación, reconocimiento de patrones o análisis semántico.

## Diferencia entre Stemming y Lematización

> El *Stemming* reduce palabras eliminando prefijos o sufijos mediante reglas simples, generando una raíz aproximada que no necesariamente existe como palabra real (por ejemplo, reduciendo "comiendo" a "com-"). La *Lematización* realiza un análisis lingüístico más completo para transformar una palabra a su forma canónica o lema del diccionario (por ejemplo, mapeando "fueron" al verbo "ser").

## Conflicto de los modelos “Caja Negra” con el diseño ético

> **¿Por qué se afirma formalmente que los modelos denominados como "Caja Negra" (Black Box) entran en conflicto con los principios del diseño ético de algoritmos?**  
> Porque su funcionamiento interno es opaco y resulta difícil comprender cómo generan sus predicciones. Esto limita la transparencia, dificulta detectar sesgos, impide auditar decisiones y complica justificar resultados erróneos o discriminatorios, entrando en conflicto con principios éticos de responsabilidad y explicabilidad.

## Definición de “Token” y tipos de tokenización

> **Define qué es un "Token" en el contexto de Procesamiento de Lenguaje Natural y proporciona ejemplos de tipos de tokenización.**  
> Un *token* es la unidad mínima de información en la que se divide un texto para poder analizarlo computacionalmente. Dependiendo de la técnica utilizada, un token puede ser una palabra (*“Me gusta la IA”* → *“Me”*, *“gusta”*, *“la”*, *“IA”*), una parte de una palabra (*“inteligencia”* → *“intel”*, *“igencia”*) o una oración completa (*“Hola, ¿cómo estás?”* → *“Hola”*, *“¿cómo estás?”*).

##  Beneficencia y No Maleficencia en IA

> La *Beneficencia* establece que los sistemas de IA deben diseñarse para generar beneficios a las personas y a la sociedad, promoviendo el bienestar, la dignidad humana y el desarrollo sostenible. La *No Maleficencia* obliga a evitar causar daño, por lo que los desarrolladores deben prevenir riesgos relacionados con la privacidad, la seguridad, la discriminación y el uso indebido de la tecnología.

## Propósito de la Visión Computacional (enfoque biológico y digital)

> Desde el *enfoque biológico*, la Visión Computacional busca comprender y modelar computacionalmente el funcionamiento del sistema visual humano, incluyendo estructuras como la retina y la corteza visual. Desde el *enfoque digital*, busca desarrollar sistemas capaces de analizar, procesar e interpretar automáticamente imágenes y videos para extraer información útil del entorno.

## Responsabilidad central del profesional informático según la ACM 2018

> El código ético de la Association for Computing Machinery (ACM) establece que el bienestar social y el interés público deben ser la máxima prioridad en el desarrollo tecnológico. Los profesionales tienen la responsabilidad de crear sistemas que respeten la dignidad humana, eviten la discriminación, protejan la privacidad y generen beneficios para la sociedad.

## Reducción de sesgo mediante optimización matemática

> La reducción del sesgo puede lograrse incorporando restricciones de equidad o penalizaciones dentro de la función de pérdida del modelo. Durante el entrenamiento, el algoritmo no solo busca minimizar errores de predicción, sino también minimizar diferencias injustas entre grupos de personas. De esta manera, el modelo es penalizado cuando genera resultados sistemáticamente desfavorables para determinados grupos, obligándolo a producir decisiones más equilibradas y justas.

## “Whole Brain Emulation” en AGI

> *Whole Brain Emulation* es una línea de investigación dentro de la Artificial General Intelligence (AGI) que busca reproducir digitalmente la estructura y funcionamiento completo del cerebro humano. Su objetivo teórico es copiar las neuronas, conexiones sinápticas, memoria y estado mental de una persona en un sistema computacional capaz de replicar funciones cognitivas avanzadas.

## Conflicto entre grandes modelos de IA y el ODS 13 (Acción por el Clima)

> Los modelos avanzados de IA requieren enormes cantidades de procesamiento para entrenarse y operar. Esto implica un alto consumo de energía en centros de datos, lo que puede generar una importante huella de carbono si la electricidad proviene de fuentes no renovables. Por ello, aunque la IA aporta muchos beneficios, también puede dificultar el cumplimiento de las metas de reducción de emisiones planteadas por el ODS 13.

## IA como motor para ODS 3 (Salud) y ODS 4 (Educación)

> En el ámbito de la *salud*, la IA ayuda a detectar enfermedades más rápidamente, apoyar diagnósticos médicos y personalizar tratamientos. En *educación*, permite crear plataformas adaptativas que identifican las necesidades de cada estudiante y ofrecen contenidos ajustados a su ritmo de aprendizaje. De esta forma mejora el acceso a servicios de salud y educación de calidad.

## Dilema ético de la automatización respecto al ODS 8 (Trabajo Decente)

> El principal dilema es que la automatización puede aumentar la productividad, pero también reemplazar ciertos empleos humanos. Desde una perspectiva ética, el objetivo no debe ser sustituir completamente a las personas, sino desarrollar sistemas que colaboren con los trabajadores, mejoren sus capacidades y promuevan la capacitación continua para evitar la exclusión laboral.

## Fusión temprana (Early Fusion) vs fusión tardía (Late Fusion)

> La *fusión temprana* combina diferentes tipos de datos, como texto, imágenes o audio, desde las primeras etapas del procesamiento para que el modelo aprenda de toda la información al mismo tiempo. La *fusión tardía* procesa cada modalidad por separado y combina las predicciones finales al final del proceso. El principal desafío de la fusión temprana es integrar correctamente datos muy diferentes entre sí, como píxeles de imágenes y palabras de texto, sin perder información importante.

## Sistemas multimodales como acercamiento a la percepción humana

> **¿Por qué se considera que los sistemas multimodales representan un acercamiento más fiel a la percepción biológica o cognitiva humana en comparación con los modelos unimodales?**  
> Porque los seres humanos interpretamos el mundo utilizando múltiples fuentes de información al mismo tiempo, como visión, audición y lenguaje. Los modelos multimodales imitan este proceso integrando diferentes modalidades de datos, lo que les permite comprender mejor el contexto y realizar tareas más complejas que los modelos unimodales.
