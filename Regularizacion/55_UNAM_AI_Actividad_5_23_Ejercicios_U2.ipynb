# Ejercicio 1. Cálculo de Entropía

Dataset:

|   Clase  | Conteo |
|----------|--------|
| Sí       | 6      |
| No       | 4      |

Calcular:

1. Entropía del nodo.



Fórmula de Entropía:

$$ H(S) = - Σ pi log2(pi) $$

<br>

Entropía = ¿Qué tan mezcladas están las clases?

Mientras más cerca de 1 → más mezclado.

Mientras más cerca de 0 → más puro.


pi es la probabilidad de que una instancia pertenezca a la clase i.

Es la proporción:

pi = fi / N

donde:

- fi = frecuencia de conteo de la clase
- N = suma total de conteos

<br>

Desarrollo Ejercicio:

- Total N = 6 + 4 = 10
- P(Sí) = 6/10 = 0.6
- P(No) = 4/10 = 0.4

<br>

H(S) = - [ P(Sí) log2(P(Sí)) + P(No) log2(P(No)) ]

= - [ 0.6 log2(0.6) + 0.4 log2(0.4) ]

= - [ (0.6)(-0.7369) + (0.4)(-1.3219) ]

= - [ -0.4421 + -0.5287 ]

= - [ -0.9708 ]

= 0.9708

<br>

Interpretación:

Como el valor está muy cerca de 1 (0.9708), el nodo tiene una impureza alta.

Si tuviéramos 5 "sí" y 5 "no", la entropía sería exactamente 1 (máxima incertidumbre).

El nodo está bastante mezclado, todavía existe mucha incertidumbre.

# Ejercicio 3. Random Forest. Bagging conceptual.

Random Forest es un algoritmo de aprendizaje supervisado basado en **Ensemble Learning**. En lugar de utilizar un solo árbol de decisión, construye muchos árboles diferentes y combina sus resultados mediante votación mayoritaria. Esto permite obtener predicciones más estables, robustas y menos sensibles al ruido.

Pregunta:

Tienes 5 árboles que predicen:

- [1, 0, 1, 1, 0]

Calcular:

1. Predicción final.

2. ¿Por qué reduce varianza?

<br>

-----

<br>

1. Predicción Final

Predicciones individuales:

```text

{1, 0, 1, 1, 0}

```

Conteo de votos:

- Clase 1: 3 votos

- Clase 0: 2 votos

Resultado:

Como **3 > 2**, la predicción final del ensamble es:

```text

Clase = 1

```

Random Forest utiliza el mecanismo de **votación mayoritaria**, por lo que la clase con más votos es la seleccionada.

<br>

2. ¿Por qué reduce varianza?

La varianza mide qué tan inestable es un modelo ante pequeños cambios en los datos.

Random Forest reduce la varianza porque combina las predicciones de muchos árboles mediante votación mayoritaria. Gracias a la Ley de los Grandes Números, los errores individuales tienden a compensarse.

Además, cada árbol se entrena con diferentes muestras de datos (Bootstrap) y diferentes subconjuntos de características, generando diversidad entre los árboles.

Este efecto, conocido como **"sabiduría de las masas"**, produce predicciones más estables y menos sensibles al ruido que las de un árbol individual.

# Ejercicio 4. Overfitting

Escenario

Un árbol tiene profundidad 20.

Calcular:

1. ¿Qué problema existe?
2. ¿Cómo lo soluciona Random Forest?
3. ¿Qué pasa si todos los árboles son iguales?

<br>

------

<br>

1. ¿Qué problema existe?

Un árbol con profundidad 20 es demasiado complejo y presenta **overfitting** (memoriza en lugar de aprender).

En lugar de aprender patrones generales, comienza a memorizar el ruido y los casos particulares del conjunto de entrenamiento.

Como consecuencia, obtiene muy buen desempeño en los datos de entrenamiento, pero falla cuando recibe datos nuevos porque tiene poca capacidad de generalización.

<br>

2. ¿Cómo lo soluciona Random Forest?

Random Forest reduce el overfitting combinando muchos árboles diferentes en lugar de depender de uno solo.

Para lograr diversidad utiliza dos mecanismos principales:

- **Bagging (Bootstrap Aggregating):** Cada árbol se entrena con una muestra aleatoria distinta de los datos.

- **Feature Randomness:** En cada división del árbol se selecciona aleatoriamente un subconjunto de atributos.

Gracias a esta diversidad, los errores de algunos árboles pueden ser compensados por los aciertos de otros, produciendo un modelo más estable y con menor varianza.

<br>

3. ¿Qué pasa si todos los árboles son iguales?

Si todos los árboles son idénticos, desaparece la diversidad del bosque.

Todos cometerán los mismos errores y la votación mayoritaria no aportará información adicional.

En ese caso no se reduce la varianza y Random Forest pierde su principal ventaja (la "sabiduría de las masas"), comportándose prácticamente como un único árbol repetido muchas veces.

# Ejercicio 5. Boosting paso a paso

Modelo inicial predice:

- y_pred = [10, 10, 10]

Valores reales:

- y = [12, 8, 15]

Calcular:

1. Calcular errores (residuos).

2. Entrenar "conceptualmente" el siguiente modelo.

3. Nueva predicción (idea general).

<br>

----

<br>

1. Calcular errores (residuos)

En Boosting, los residuos son la diferencia entre el valor real y la predicción actual.

Representan "lo que el modelo aún no explica".

La fórmula es:

$$ Residuo = y - y_pred $$


Cálculos:

- Para y₁: 12 - 10 = 2

- Para y₂: 8 - 10 = -2

- Para y₃: 15 - 10 = 5

Vector de residuos: **[2, -2, 5]**

<br>

2. Entrenar conceptualmente el siguiente modelo

El segundo modelo se entrena utilizando los residuos generados por el primero.

Su objetivo es aprender el patrón de errores: ***[2, -2, 5]***

para corregir las predicciones que aún están alejadas de los valores reales.

El nuevo modelo intentará entender:

- Por qué faltaron 2 unidades en el primer caso.

- Por qué nos pasamos por 2 unidades en el segundo caso.

- Por qué faltaron 5 unidades en el tercer caso.

<br>

3. Nueva predicción (idea general)

La nueva predicción se obtiene sumando una fracción de la corrección aprendida por el segundo árbol.

En cada iteración las predicciones se acercan gradualmente a los valores reales y el error global disminuye.

Idea general:

```text

Nueva predicción = Predicción anterior + Corrección aprendida

```

Por eso Gradient Boosting construye modelos secuencialmente: cada nuevo árbol intenta corregir los errores cometidos por los árboles anteriores.

# Ejercicio 6. Diferencia Gradient Boosting vs Random Forest

Pregunta conceptual

Calcular:

1. Diferencia entre:
   - Random Forest
   - Gradient Boosting
2. ¿Cuál sobreajusta más fácilmente?
3. ¿Cuál es más interpretable?

<br>

------

<br>

1. Diferencia entre Random Forest y Gradient Boosting

La diferencia principal es la forma en la que construyen los árboles.


Random Forest crea muchos árboles independientes en paralelo. Cada árbol genera una predicción y la respuesta final se obtiene mediante votación (clasificación) o promedio (regresión). Su objetivo principal es reducir la varianza y evitar el sobreajuste.

Gradient Boosting construye los árboles de forma secuencial. Cada nuevo árbol intenta corregir los errores o residuos cometidos por los árboles anteriores. Su objetivo principal es reducir el sesgo y mejorar progresivamente la precisión del modelo.

Comparación resumida

| Característica  | Random Forest | Gradient Boosting |
|----------|----------|--------|
| Construcción | Árboles en paralelo    | Árboles secuenciales     |
| Objetivo | Reducir varianza     | Reducir sesgo     |
| Mecanismo | Votación o promedio     | Corrección de errores     |
| Robustez| Más robusto al ruido    | Más sensible al ruido     |
| Riesgo de overfitting| Menor    | Mayor     |

<br>

2. ¿Cuál sobreajusta más fácilmente?

Gradient Boosting sobreajusta más fácilmente porque intenta corregir continuamente los errores del entrenamiento. Si se agregan demasiados árboles o el modelo es muy complejo, puede comenzar a aprender ruido en lugar de patrones reales. Random Forest reduce este riesgo gracias al uso de aleatoriedad, Bootstrap y votación entre múltiples árboles.

<br>

3. ¿Cuál es más interpretable?

Random Forest suele ser más interpretable porque permite analizar con mayor facilidad la importancia de las variables y el comportamiento promedio del bosque.

En Gradient Boosting la interpretación es más compleja debido a que cada árbol depende de los errores corregidos por los árboles anteriores.

# Ejercicio 7: K-Means – Iteración manual

## Datos
Puntos:  
(1,1), (2,2), (8,8), (9,9)

Centroides iniciales:  
- C1 = (1,1)  
- C2 = (9,9)

## 1. Asignación de clusters (distancia euclidiana)

$$ d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2} $$

- (1,1): distancia a C1 = 0, distancia a C2 = $\sqrt{(9-1)^2+(9-1)^2} = \sqrt{128} \approx 11.31$ → **Cluster 1**
- (2,2): a C1 = $\sqrt{(2-1)^2+(2-1)^2} = \sqrt{2} \approx 1.41$; a C2 = $\sqrt{(9-2)^2+(9-2)^2} = \sqrt{98} \approx 9.90$ → **Cluster 1**
- (8,8): a C1 = $\sqrt{(8-1)^2+(8-1)^2} = \sqrt{98} \approx 9.90$; a C2 = $\sqrt{(9-8)^2+(9-8)^2} = \sqrt{2} \approx 1.41$ → **Cluster 2**
- (9,9): a C1 = $\sqrt{128} \approx 11.31$; a C2 = 0 → **Cluster 2**

**Clusters resultantes:**  
G1 = {(1,1), (2,2)}  
G2 = {(8,8), (9,9)}

## 2. Recalcular centroides (media)

Nuevo C1 = $\left( \frac{1+2}{2}, \frac{1+2}{2} \right) = (1.5, 1.5)$

Nuevo C2 = $\left( \frac{8+9}{2}, \frac{8+9}{2} \right) = (8.5, 8.5)$

## 3. Una iteración más (¿converge?)

Con centroides (1.5,1.5) y (8.5,8.5):

- (1,1): distancia a C1 = $\sqrt{(1-1.5)^2+(1-1.5)^2} = \sqrt{0.5} \approx 0.71$; a C2 = $\sqrt{(8.5-1)^2+(8.5-1)^2} = \sqrt{112.5} \approx 10.61$ → sigue en G1
- (2,2): a C1 = $\sqrt{(2-1.5)^2+(2-1.5)^2} = 0.71$; a C2 = $\sqrt{(8.5-2)^2+(8.5-2)^2} = \sqrt{84.5} \approx 9.19$ → sigue en G1
- (8,8): a C1 = $\sqrt{(8-1.5)^2+(8-1.5)^2} = \sqrt{84.5} \approx 9.19$; a C2 = $\sqrt{(8.5-8)^2+(8.5-8)^2} = 0.71$ → sigue en G2
- (9,9): a C1 = $\sqrt{(9-1.5)^2+(9-1.5)^2} = \sqrt{112.5} \approx 10.61$; a C2 = 0.71 → sigue en G2

No hay cambios en la asignación → **los centroides no cambian más**. El algoritmo converge.

````python
# Ejercicio 7: K-Means manual
import numpy as np

# Datos
puntos = np.array([(1,1), (2,2), (8,8), (9,9)])
centroides = np.array([(1,1), (9,9)])

def distancia(a, b):
    return np.sqrt((a[0]-b[0])**2 + (a[1]-b[1])**2)

# Iteración 1
print("--- Iteración 1 ---")
clusters = {0: [], 1: []}
for p in puntos:
    dists = [distancia(p, c) for c in centroides]
    cluster = np.argmin(dists)
    clusters[cluster].append(p)
    print(f"Punto {p} -> Cluster {cluster+1} (distancias: C1={dists[0]:.2f}, C2={dists[1]:.2f})")

# Recalcular centroides
nuevos = []
for i in range(2):
    if clusters[i]:
        nuevo = np.mean(clusters[i], axis=0)
        nuevos.append(nuevo)
    else:
        nuevos.append(centroides[i])
nuevos = np.array(nuevos)
print(f"\nNuevos centroides: C1={nuevos[0]}, C2={nuevos[1]}")

# Iteración 2
print("\n--- Iteración 2 ---")
clusters2 = {0: [], 1: []}
for p in puntos:
    dists = [distancia(p, c) for c in nuevos]
    cluster = np.argmin(dists)
    clusters2[cluster].append(p)
    print(f"Punto {p} -> Cluster {cluster+1} (distancias: C1={dists[0]:.2f}, C2={dists[1]:.2f})")

# Verificar convergencia
if np.array_equal(nuevos, centroides):
    print("\nLos centroides no cambiaron → convergencia alcanzada.")
else:
    print("\nAún hay cambios. Se necesitarían más iteraciones.")
````

# Ejercicio 8: K-Means - Elección de K en K-Means

## 1. ¿Qué pasa si K es muy pequeño? (ejemplo K=1 o mucho menor al número natural de grupos)

- **Sobajuste (underfitting):** El modelo agrupa puntos que en realidad pertenecen a clusters diferentes dentro de un mismo grupo.
- **Inercia (WCSS) alta:** La suma de las distancias al cuadrado de cada punto a su centroide es muy grande, porque los puntos están lejos del centro único.
- **Pérdida de estructura:** No se capturan los patrones naturales de los datos.

## 2. ¿Qué pasa si K es muy grande? (ejemplo K = N, donde N es número de puntos)

- **Sobreajuste (overfitting):** Cada punto forma su propio cluster.
- **Fragmentación excesiva:** La inercia total es 0, pero el modelo no tiene valor práctico porque no generaliza. Se pierde el propósito del clustering.
- **Alto costo computacional** y difícil interpretación.

## 3. Método del codo (Elbow Method)

El método del codo busca el valor de K donde la reducción de la inercia (WCSS – Within-Cluster Sum of Squares) deja de ser significativa, es decir, el punto de equilibrio entre simplicidad y ajuste.

**Procedimiento:**
1. Calcular la inercia para K = 1, 2, 3, …, hasta un valor máximo.
2. Graficar K vs inercia.
3. Buscar el “codo” – el punto donde la curva empieza a aplanarse.

**Ejemplo numérico:**

| K (clusters) | Inercia (WCSS) | Reducción de inercia |
|--------------|----------------|------------------------|
| 1            | 500            | –                      |
| 2            | 250            | 250                    |
| 3            | 100            | 150   Centro de reflexion                 |
| 4            | 85             | 15                     |
| 5            | 75             | 10                     |

**Análisis:**
- De K=1 a K=2 la inercia baja drásticamente (250 unidades).
- De K=2 a K=3 baja todavía 150 unidades.
- De K=3 a K=4 la reducción es mínima (solo 15 unidades).
- Por lo tanto, **K=3 es el punto de codo**; aumentar más clusters no vale la pena porque la ganancia en inercia es muy pequeña.

**Fórmula de inercia (WCSS):**

$$ \text{WCSS} = \sum_{i=1}^{K} \sum_{x \in C_i} \|x - \mu_i\|^2 $$

donde $C_i$ es el cluster $i$ y $\mu_i$ su centroide.

````python
# Ejercicio 8: Método del codo (Elbow Method)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generar datos sintéticos con 3 clusters naturales
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.8, random_state=42)

# Calcular inercia para K de 1 a 10
inercia = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inercia.append(kmeans.inertia_)

# Mostrar tabla de inercias
print("K\tInercia (WCSS)\tReducción")
prev = None
for i, k in enumerate(K_range):
    reduc = "" if i == 0 else f"{prev - inercia[i]:.2f}"
    print(f"{k}\t{inercia[i]:.2f}\t\t{reduc}")
    prev = inercia[i]

# Graficar el codo
plt.figure(figsize=(8,5))
plt.plot(K_range, inercia, 'bo-', markersize=8)
plt.xlabel('Número de clusters (K)')
plt.ylabel('Inercia (WCSS)')
plt.title('Método del codo para seleccionar K')
plt.xticks(K_range)
plt.grid(True)
plt.show()

print("\nInterpretación: El codo se forma aproximadamente en K=3,")
print("lo que coincide con los clusters reales de los datos generados.")
````

### Punto A (1,1)

- Distancia a C: $\sqrt{(2-2)^2 + (2-1)^2} = 1$  
- Distancia a D: $\sqrt{(8-2)^2 + (8-1)^2} = \sqrt{85} \approx 9.220$  

Vecinos con distancia ≤ 2: A, B, C → **3 puntos**.

### Punto C (2,2)
- Distancia a A: 1.414  
- Distancia a B: 1  
- Distancia a C: 0  
- Distancia a D: $\sqrt{(8-2)^2 + (8-2)^2} = \sqrt{72} \approx 8.485$  

Vecinos con distancia ≤ 2: A, B, C → **3 puntos**.

### Punto D (8,8)
- Distancia a A: 9.899  
- Distancia a B: 9.220  
- Distancia a C: 8.485  
- Distancia a D: 0  

Vecinos con distancia ≤ 2: solo D (él mismo) → **1 punto**.

## 2. Clasificación según DBSCAN

- **Punto núcleo (core)**: tiene al menos `minPts = 3` puntos en su vecindad (incluyéndose).  
  → A, B, C son **núcleo**.

- **Punto borde (border)**: tiene menos de `minPts` vecinos, pero es alcanzable desde un núcleo (está a distancia ≤ eps de al menos un núcleo).  
  → En este caso, todos los vecinos de los núcleos ya son núcleos. No hay puntos borde.

- **Punto ruido (noise)**: no es núcleo ni está a distancia ≤ eps de ningún núcleo.  
  → D (8,8) es **ruido**, porque su única vecindad es él mismo y está lejos de los demás.

## 3. Respuesta final
- **Núcleo**: (1,1), (2,1), (2,2)  
- **Borde**: ninguno  
- **Ruido**: (8,8)

## Nota conceptual
A diferencia de K‑means, DBSCAN no fuerza a todos los puntos a pertenecer a un cluster. Los puntos aislados se detectan automáticamente como ruido. En K-Means el (8,8) hubiera sido forzado a pertenecer a un grupo.

Aquí no hay centros, hay vecindades. Un punto se incluye a sí mismo en el conteo para su vecindad.

````python
# Ejercicio 9: Clasificación DBSCAN (manual)
import numpy as np
from sklearn.metrics import pairwise_distances

# Datos
puntos = np.array([(1,1), (2,1), (2,2), (8,8)])
nombres = ['A (1,1)', 'B (2,1)', 'C (2,2)', 'D (8,8)']

eps = 2
minPts = 3

# Matriz de distancias
distancias = pairwise_distances(puntos, metric='euclidean')
np.set_printoptions(precision=3, suppress=True)
print("Matriz de distancias:")
print(distancias)
print()

# 1. Identificar vecinos dentro de eps (incluyéndose a sí mismo)
vecinos = []
for i in range(len(puntos)):
    vecinos_i = [j for j in range(len(puntos)) if distancias[i,j] <= eps]
    vecinos.append(vecinos_i)
    print(f"{nombres[i]} -> vecinos (índices): {vecinos_i}, total = {len(vecinos_i)}")

# 2. Clasificar núcleos (core)
core = [i for i in range(len(puntos)) if len(vecinos[i]) >= minPts]
print(f"\nPuntos núcleo (core): {[nombres[i] for i in core]}")

# 3. Identificar bordes y ruido
# Un punto es borde si no es núcleo pero es alcanzable desde algún núcleo (distancia <= eps)
border = []
noise = []
for i in range(len(puntos)):
    if i in core:
        continue
    # Es alcanzable desde algún núcleo?
    alcanzable = any(distancias[i, c] <= eps for c in core)
    if alcanzable:
        border.append(i)
    else:
        noise.append(i)

print(f"Puntos borde: {[nombres[i] for i in border]}")
print(f"Puntos ruido: {[nombres[i] for i in noise]}")

# Verificación con DBSCAN de sklearn (opcional)
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=eps, min_samples=minPts).fit(puntos)
print("\n--- Verificación con sklearn ---")
for i, label in enumerate(db.labels_):
    # label = -1 es ruido, 0,1,... son clusters
    tipo = "ruido" if label == -1 else f"cluster {label}"
    print(f"{nombres[i]} -> {tipo}")
````

# Ejercicio 10: Detección de Anomalías – Isolation Forest

## Dataset
[10, 12, 11, 13, 100]

## 1. Identificar la anomalía mediante Z‑score

### Media ($\bar{x}$)
$$
\bar{x} = \frac{10 + 12 + 11 + 13 + 100}{5} = \frac{146}{5} = 29.2
$$

### Varianza ($\sigma^2$) y Desviación estándar ($\sigma$)

Diferencias al cuadrado:
- $(10 - 29.2)^2 = 368.64$
- $(12 - 29.2)^2 = 295.84$
- $(11 - 29.2)^2 = 331.24$
- $(13 - 29.2)^2 = 262.44$
- $(100 - 29.2)^2 = 5012.64$

Suma: $368.64 + 295.84 + 331.24 + 262.44 + 5012.64 = 6270.8$

$$
\sigma^2 = \frac{6270.8}{5} = 1254.16
$$
$$
\sigma = \sqrt{1254.16} \approx 35.41
$$

### Z‑score del punto 100
$$
Z = \frac{100 - 29.2}{35.41} \approx 2.00
$$

En una distribución normal, un valor con $|Z| > 1.96$ se considera inusual (anomalía).  
Aunque el dataset es muy pequeño, el punto **100** es claramente el candidato a anomalía. Los otros puntos tienen Z‑score cercano a 0 o negativo, mientras que 100 tiene Z ≈ 2.0.

## 2. ¿Por qué Isolation Forest lo detecta rápido?

Isolation Forest se basa en particiones aleatorias recursivas:
- Las anomalías son **pocas y diferentes** → se aíslan con muy pocos cortes aleatorios.
- El punto 100 está muy lejos del grupo principal, por lo que es muy probable que una partición aleatoria lo separe inmediatamente del resto.

> **Intuición:** Es más fácil aislar a una persona sola en un campo que a una persona en medio de una multitud.

## 3. ¿Qué significa “profundidad corta”?

- **Profundidad (path length):** Número de particiones (niveles del árbol) necesarias para aislar completamente un punto.
- **Profundidad corta:** El punto se aísla en muy pocas particiones → indicador fuerte de anomalía.
- **Profundidad larga:** El punto está en una región densa y necesita muchas particiones para separarse → indicador de punto normal.

En nuestro caso, el 100 tendrá una profundidad corta, mientras que 10, 11, 12, 13 tendrán profundidad larga.

````python
# Ejercicio 10: Detección de anomalías con Z‑score e Isolation Forest
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest

# Datos
datos = np.array([10, 12, 11, 13, 100]).reshape(-1, 1)
nombres = ['10', '12', '11', '13', '100']

# 1. Cálculo manual de Z‑score
media = np.mean(datos)
std = np.std(datos, ddof=0)  # población
z_scores = (datos - media) / std

print("--- Z‑score (detección estadística) ---")
print(f"Media = {media:.2f}, Desviación estándar = {std:.2f}\n")
for i, z in enumerate(z_scores):
    print(f"Punto {nombres[i]}: Z = {z[0]:.4f}")

print("\nEl punto 100 tiene Z ≈ 2.0, mientras los demás están cerca de -0.5. Es la anomalía.")

# 2. Isolation Forest
iso_forest = IsolationForest(contamination=0.2, random_state=42, n_estimators=100)
iso_forest.fit(datos)
predicciones = iso_forest.predict(datos)
# -1 = anomalía, 1 = normal

# Obtener las profundidades (path lengths) de cada muestra
# En IsolationForest, score_samples da el negativo de la profundidad media
scores = iso_forest.score_samples(datos)  # más negativo = más anómalo
# Para convertir a profundidad (a mayor profundidad, más normal)
# Podemos normalizar: la profundidad media es -score_samples (aproximadamente)
path_lengths = -scores

print("\n--- Isolation Forest ---")
for i, (nom, pred, pl) in enumerate(zip(nombres, predicciones, path_lengths)):
    tipo = "Anomalía" if pred == -1 else "Normal"
    print(f"Punto {nom}: {tipo} | Profundidad (path length) ≈ {pl:.2f}")
    if nom == '100':
        print("  → Profundidad corta (menor que las demás) → se aísla rápido.")
    else:
        print("  → Profundidad más larga → punto normal.")

# Verificación gráfica de la profundidad
print("\n--- Explicación de profundidad ---")
print("El Isolation Forest entrena múltiples árboles. Para el punto 100,")
print("la mayoría de los árboles lo aíslan en pocas particiones (profundidad corta),")
print("mientras que para los puntos cercanos (10,11,12,13) se necesitan muchas más particiones.")
````
