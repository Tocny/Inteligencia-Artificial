#### Por palabras
Usualmente los textos se dividen en palabras, algunas otras veces en sus morfemas (léxicos y gramaticales).


La forma más sencilla de partir un texto en palabras es al "dividir" usando los espacios en blanco.

```python
frase = 'El puerquito tiene sombrero de copa blanco, un monóculo, y un elegante smoking'
print('{0:15}'.format('--Token--'))
for token in frase.split(' '):
  print(token)
```

Como podemos observar, las comas siguen "pegadas" a sus palabras, para evitar esto, podemos usar tokenizadores como el de NLTK.

```python
from nltk.tokenize import NLTKWordTokenizer
frase = 'El puerquito tiene sombrero de copa blanco, un monóculo, y un elegante smoking'
tokenizer = NLTKWordTokenizer()
print('{0:15}'.format('--Token--'))
for token in tokenizer.tokenize(frase):
  print(token)
```

### El preprocesamiento

En general, para cualquier tarea de aprendizaje automático, dígase procesamiento de señales, procesamiento de imágenes, necesitamos remover toda esta basura o paja que solo mete ruido a nuestros datos. En el PLN es escencial que hagamos un correcto preprocesamiento de los datos para mejorar la calidad de los mismos.

#### ¿En qué consiste el preprocesamientooooo?
Se trata de remover entidades textuales y editar las existentes para llevar nuestro texto a su forma formal o canónica en lo posible, tenemos que pasar del lenguaje enriquecido por variaciones en el mismo a su forma formal. Para lograr esto, tenemos que lidiar con 5 principales fuentes de ruido:
- Entidades especiales
- Metamorfosis de las palabras
- Palabras auxiliares
- Segmentación
- Normalización

De a poco iremos descubriendo qué es cada una de estas "cosas"

#### Entidades especiales

Son entidades embebidas en el texto, cuya función no es importante para el análisis del mismo, si no más bien para otras aplicaciones como los delimitadores HTML lo son para HTML. Algunas de las entidades especiales más comunes en los textos son:
- URLs
- Entidades HTML
- Entidades propias de la web que consultamos (redes sociales como X tiene el @User, o los #Hashtag)
- Emojis 🍆 🍑 👀
- Entidades no ASCII

##### Pidámosle ayuda a Códex 🦸, para enfrentar a las malevolas entidades

- URLs: "Texto...https://www..." --> Códex --> "Texto..."
- Entidades HTML: "\<HTML>Texto...<\/HTML>" --> Códex --> "Texto..."
- "Texto... @User RT:Texto..." --> Códex --> "Texto... Texto..."
- "Texto... @User RT:Texto..." --> Códex --> "Texto... \_USER\_ \_RT\_ Texto..."
- 🍆 🍑 👀 --> Códex --> ""
- 🍆 🍑 👀 --> Códex --> "berenjena durazno ojos"
- 🍆 🍑 👀 --> Códex --> "Inserte su mejor traducción no literal"$^*$
- Entidades no ASCII --> ""

#### Metamorfosis de las palabras
Son cambios que sufre el lenguaje formal debido a modismos o popularismos, entre otros usos del lenguaje.

- Fonética
- Morfológica
- Semántica

Estos cambios se puedes presentar a manera de:
- Abreviaciones: "Atte." ,"Sr.", "Sra.", "cap", "pá", "má"
- Elongaciones: "Buenaaaaaaas", "Nooooooo", "Holaaaaa"
- Transformaciones: "ntp", "GPI", "chi", "shi", "ño", "nop", "simon", "nel", "alv", "grax", "chtm", "ptm"
- Combinaciones: Holiiiiii, naaaaaa, simoooon, ñoñoño

##### Códeeeeeeex, ayuraaaaa

Para lidiar con estos cambios morfológicos, podemos usar:
- Lexicones
- Normalizar las repeticiones a solo dos apariciones
- Eliminarlos

Ejemplo mediante lexicones y normalización:

- "Holiiiiiiii" --> "Holii" --> "Hola"

#### Palabras auxiliares (*stop words*)

Son palabras como los artículos, pronombres o preprosiciones, o de cualquier otra índole, cuya frecuencia de aparición en un documento es muy alta y puede afectar la tarea que queremos resolver (nótese que para estilos de redacción estas palabras son muy necesarias)

#### Segmentación
- Stemización
- Lematización

#### Normalización

- Transformamos mayúsculas a minúsculas o viceversa
- Quitamos los acentos
- Expandimos las contracciones
- Removemos la puntuación
- L33t 5p3ak1ng

### Convirtiendo texto a valores numéricos (vectores)

Representar palabras a través de números es una tarea sencilla, pero llega una problemática al querer representar la semántica, morfología, sintaxis e incluso el contexto de lo que estamos expresando en un texto a través de números. ¿Cómo lo representamos?

Lo primero en lo que se pensó fue en ¿qué hacer con las palabras repetidas?

Para esta sección utilicemos como ejemplo el texto ya preprocesado del post de X

```python
x_post_prep = post_from_x_no_alpha
print(x_post_prep)
```

#### El vocabulario

El vocabulario es el conjunto de palabras únicas que contienen todos los textos dentro de nuestra colección. Por medio de este, evitamos palabras repetidas en los textos, y reducimos la cantidad de texto a vectorizar.

```python
# El vocabulario
vocab = list(set(x_post_prep.split()))
print(vocab)
```

#### Asignaciones aleatorias

Las asignaciones aleatorias consisten en asignar un número aleatorio a cada palabra del vocabulario, el problema con esta representación es que podemos tener números repetidos.


```python
import random

random_vec = random.sample(population= range(100), k= 27)
print(f'Tamaño del vocabulario: {len(vocab)}')
print(f'Tamaño del vector: {len(random_vec)}')
print(dict(zip(vocab, random_vec)))
```

#### Representación one-hot a nivel token

Para hacer este tipo de representaciones, creamos un vector de ceros del tamaño del vocabulario, y si la palabra aparece, ponemos un uno en la posición i-ésima del vector, correspondiente a la palabra en el orden del vocabulario

```python
# Calculamos la frecuencia de aparición de cada palabra del vocabulario en el texto
frequencies = []
for i, token in enumerate(vocab):
  # Creamos la representacion one_hot
  one_hot = [0 for i in range(len(vocab))]

  if x_post_prep.count(token) > 0:
    one_hot[i] = 1

  frequencies.append(one_hot)

# Creamos un diccionario donde tengamos el vocabulario y su representación one hot para nuestros textos
token_one_hot = dict(zip(vocab, frequencies))
print(token_one_hot)
```

#### Problemas con este tipo de representaciones

Este tipo de representaciones consumen una cantidad de memoria muy alta, ya que para representar un texto, necesitariamos una matriz de tamaño:

$N_p \times L_V$

Por ejemplo, si nuestro corpus tiene 100 textos y cada texto tiene en promedio 40 palabras y el vocabulario de este corpus es de 500 palabras, necesitaríamos de una matriz de $(100 \times 40) \times 500 = 2,000,000$ de entradas. Además, estas matrices tienen poca información ya que están repletas de ceros.

#### Representación one-hot a nivel documento

Al ver que estas representaciones consumían una gran cantidad de memoria, se decidió representar un documento completo, reduciendo la cantidad de entradas a solo $100 \times 500 = 50,000$ entradas en la matriz. Para hacer esta representación, ahora cada docuemnto se engloba en un vector en el que si la palabra del documento aparece se asigna uno y si no un cero.

```python
onehot_sent = [0 for i in range(len(vocab))]
for token in vocab:
  if token in x_post_prep:
    onehot_sent[vocab.index(token)] = 1
print(onehot_sent)
```

en este caso, como el total del vocabulario aparece en el texto, nuestro vector de características estará lleno de unos. Conforme el vocabulario sea más variado entre documentos, la cantidad de ceros disminuye.

#### La bolsa de palabras
Lo primero que se consideró fue en crear una bolsa de palabras que almacene todas las palabras únicas de un conjunto de textos, a este modelo se le conoce como *Bag of Words* (BoW).

Ahora basta con calcular la frecuencia de cada una de las palabras en nuestro texto

```python
# Calculamos la frecuencia de aparición de cada palabra del vocabulario en el texto
frequencies = []
for token in vocab:
  freq = x_post_prep.count(token)
  frequencies.append(freq)

# Creamos un diccionario donde tengamos el vocabulario y su frecuencia para nuestros textos
bow = dict(zip(vocab, frequencies))
print(bow)
```

#### Frecuencia de términos (TF)
De la bolsa de palabras podemos obtener la representación vectorial del texto mejor conocida como representación por frecuencia de términos o por conteos.

```python
# Representación por Frecuencia de Términos (TF)
print(frequencies)
```

#### TF-IDF
La representación TF-IDF (frecuencia de término-frecuencia inversa de documento), evalúa la importancia de cada una de las palabras respecto al documento (texto) en el que aparece y contra la colección de documentos del corpus que se está analizando.

Se calcula como:

$tf(t, d) = f_{t, d}$

$idf(t, D) = log \frac{N}{|\{ d \in D : t \in d \}|}$

$tf \cdot idf (t, d, D) = tf(t, d) \cdot idf(t, D)$

Dónde $t$ es el término a analizar; $d$ es el documento en el que aparece el término analizado; $D$ es la colección de documentos


Partamos nuestro texto en tres sub-cadenas de texto.

```python
texto_1 = x_post_prep.split()[:10]
texto_2 = x_post_prep.split()[7:18]
texto_3 = x_post_prep.split()[15:]
print(texto_1)
print(texto_2)
print(texto_3)
```

```python
tf_complice_texto_2 = texto_2.count('complice')
print(f'TF de complice en el texto 2: {tf_complice_texto_2}')
```

```python
import numpy as np
idf_complice = np.log(3 / (1 + (1 if 'complice' in texto_1 else 0) + (1 if 'complice' in texto_2 else 0) + (1 if 'complice' in texto_3 else 0)))
print(f'IDF de complice: {idf_complice}')
```

```python
tf_idf_complice = np.dot(tf_complice_texto_2, idf_complice)
print(tf_idf_complice)
```

Ahora, calculamos para cada palabra de cada una de las frases y obtenemos una representación vectorial de nuestro texto

```python
tf_idf_all = []
for text in [texto_1, texto_2, texto_3]:
  tfidf = []
  for token in vocab:
    tf = text.count(token)
    idf = np.log(3 / (1 + (1 if token in texto_1 else 0) + (1 if token in texto_2 else 0) + (1 if token in texto_3 else 0)))
    tfidf.append(float(tf * idf))
  tf_idf_all.append(tfidf)
print(f'TF-IDF para el texto 1:\n{tf_idf_all[0]}\n')
print(f'TF-IDF para el texto 2:\n{tf_idf_all[1]}\n')
print(f'TF-IDF para el texto 3:\n{tf_idf_all[2]}')
```

### Similitud entre textos

Una vez que tenemos nuestros textos vectorizados, podemos hacer varias cosas, una de ellas es ver que tan similar es un documento repecto a otro.
Existen varios métodos para determinar la similitud, que usualmente se refieren a distancia entre dos puntos, o similitud entre vectores. Algunos de estos son:
- [Distancia euclideana](https://es.wikipedia.org/wiki/Distancia_euclidiana)
- [Distancia de Hamming](https://en.wikipedia.org/wiki/Hamming_distance)
- [Distancia Manhattan](https://www.tusclasesparticulares.com/blog/como-calcular-distancia-entre-dos-puntos#:~:text=5%20unidades.-,F%C3%B3rmula%20de%20distancia%20de%20Manhattan,-%3A%20otra%20forma%20de)
- [Similitud conseno](https://en.wikipedia.org/wiki/Cosine_similarity)

Acontinuación veremos cómo implementar la similitud conseno y las ecuaciones matemáticas que nos ayudan a calcularla

$\text{cosine similarity} = S_C(A, B) := \frac{A.B}{||A||\; ||B||} = \frac{\sum_{i=1}^{n}A_iB_i}{\sqrt{\sum_{i=1}^nA_i^2}\cdot \sqrt{\sum_{i=1}^nB_i^2}}$

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity(np.asarray(tf_idf_all[2]).reshape(1, -1), np.asarray(tf_idf_all[1]).reshape(1, -1), dense_output= True)
```

```python
def norma(vec):
  normal = np.array([])
  suma = 0
  for i in vec:
    suma += (np.square(i))

  norma = np.sqrt(suma)
  return norma

def similitud(v1, v2):
  prod_interno = v1.dot(v2)
  norma_v1 = norma(v1)
  norma_v2 = norma(v2)
  simil_v1_v2 = prod_interno / (norma_v1 * norma_v2)

  return round(simil_v1_v2, 5)
```

```python
similitud(np.array(tf_idf_all[0]), np.array(tf_idf_all[2]))
```
