# Funciones

Esta función convierte un color en formato RGB (valores rojo, verde, azul) a su representación en formato hexadecimal

```python
def rgb_to_hex(rgb_color):
    # Inicializamos una cadena con el símbolo "#" que representa un color en formato hexadecimal.
    hex_color = "#"

    # Iteramos sobre cada valor en la tupla o lista rgb_color.
    for i in rgb_color:
        # Convertimos el valor actual (i) a un entero por si no lo es.
        i = int(i)

        # Convertimos el valor entero a su representación hexadecimal de dos dígitos.
        # El formato {:02x} asegura que siempre haya dos dígitos en hexadecimal, añadiendo un 0 si es necesario.
        hex_color += ("{:02x}".format(i))

    # Retornamos el color en formato hexadecimal con el prefijo "#".
    return hex_color
```

Esta función toma una imagen original (raw_img), la redimensiona a un tamaño más pequeño (500x200), y luego convierte la imagen en un arreglo bidimensional donde cada fila representa un píxel con sus valores RGB. Esto es útil para simplificar la manipulación de la imagen, como aplicar algoritmos de procesamiento sobre sus píxeles.

```python
def prep_image(raw_img):
    # Redimensionamos la imagen raw_img a un tamaño de 500x200 píxeles.
    # cv2.INTER_AREA es utilizado como método de interpolación, que es ideal para reducir el tamaño de la imagen.
    modified_img = cv2.resize(raw_img, (500, 200), interpolation=cv2.INTER_AREA)

    # Cambiamos la forma de la imagen redimensionada a una matriz 2D.
    # Esto convierte la imagen de un arreglo 3D (alto, ancho, canales) a un arreglo 2D, donde cada fila representa un píxel,
    # y cada fila tiene 3 valores correspondientes a los canales de color (RGB).
    modified_img = modified_img.reshape(modified_img.shape[0] * modified_img.shape[1], 3)

    # Devolvemos la imagen modificada, ahora en formato de matriz 2D.
    return modified_img
```

Esta función analiza los colores principales de una imagen utilizando KMeans para agrupar los colores en 5 clústeres (colores representativos). Luego, crea un gráfico de pastel que muestra la proporción de cada color en la imagen, guardando el gráfico en un archivo. Los colores de los clústeres se muestran en formato hexadecimal y se imprimen en la consola.

```python
def color_analysis(img):
    # Creamos un modelo de KMeans para agrupar los colores en 5 clústeres.
    clf = KMeans(n_clusters=5)

    # Ajustamos el modelo KMeans a los datos de la imagen (img) y predecimos a qué clúster pertenece cada píxel.
    color_labels = clf.fit_predict(img)

    # Obtenemos los colores centrales de cada uno de los 5 clústeres (es decir, los colores representativos).
    center_colors = clf.cluster_centers_

    # Contamos cuántos píxeles pertenecen a cada clúster (frecuencia de cada color).
    counts = Counter(color_labels)

    # Imprimimos la cantidad de píxeles asociados a cada color (clúster).
    print(counts)

    # Ordenamos los colores según los clústeres más frecuentes (los más comunes aparecen primero).
    ordered_colors = [center_colors[i] for i in counts.keys()]

    # Convertimos los colores RGB de los clústeres a formato hexadecimal.
    hex_colors = [rgb_to_hex(ordered_colors[i]) for i in counts.keys()]

    # Configuramos el tamaño de la figura para la gráfica de pastel (pie chart).
    plt.figure(figsize=(12, 8))

    # Creamos un gráfico de pastel donde cada sección representa un color y su proporción en la imagen.
    plt.pie(counts.values(), labels=hex_colors, colors=hex_colors)

    # Guardamos la gráfica de pastel como una imagen "color_analysis_report.png".
    plt.savefig("color_analysis_report.png")

    # Imprimimos los colores en formato hexadecimal.
    print(hex_colors)
```
