**Detección del punto de interés**

En procesamiento de imágenes, un atributo o característica de imagen es un fragmento de información sobre el contenido de esta. Por lo general esta información es sobre si una determinada región de la imagen contiene ciertas propiedades. Los atributos o características pueden ser estructuras específicas como puntos, bordes, esquinas u objetos en particular, así como también pueden ser el resultado de una operación hecha sobre una vecindad de puntos.

**La transformación de características invariantes de escala (SIFT)** es un algoritmo de detección en visión computacional. Permite identificar características locales dentro de una imagen conocidas como puntos de interés. Los puntos de interés son invariantes a la escala y rotación. La principal ventaja de esta transformación es que las características obtenidas no se ven afectadas por el tamaño de la imagen o su orientación, lo que significa que independientemente de si la imagen es girada o modificada en escala, se encontrarán los mismos puntos de interés.

El proceso para la identificación de estos puntos de interés se divide en cuatro etapas (Lowe, 2004):

-  **Construcción del espacio de escala**. Se identifican la ubicación y escalas de las características que se pueden reconocer de forma repetida desde diferentes perspectivas en el mismo objeto o escena. Se asegura que las características sean independientes a la escala.
-  **Localización de puntos interés.** Ajusta un modelo para determinar la ubicación y escalas de las características seleccionando puntos clave en función de una medida de estabilidad.
-  **Asignación de la orientación.** Se asegura de que los puntos de interés sean invariantes a la rotación. Se calculan las mejores orientaciones para cada región que contiene puntos de interés.
-  **Descriptor del punto de interés.** Asigna una “huella digital” única a cada punto de interés. Se utilizan degradados de imagen en la escala y rotación seleccionada para describir cada región que contiene puntos de interés.

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img1 = cv.imread('libertad.jpg')
gray1  = cv.cvtColor (img1, cv.COLOR_BGR2GRAY)

#Puntos de intes
sift =cv.SIFT_create()

kp1, descriptor1 =sift.detectAndCompute(img1, None)
img1kp = cv.drawKeypoints(gray1, kp1, img1,
                          flags = cv.DrawMatchesFlags_DRAW_RICH_KEYPOINTS)
plt.figure(figsize = (6,6))

plt.imshow(img1kp)
```

```python
img1 = cv.imread('libertad.jpg')
img1  = cv.cvtColor (img1, cv.COLOR_BGR2GRAY)
img2 = cv.imread('libertad2.jpg')
img2  = cv.cvtColor (img2, cv.COLOR_BGR2GRAY)
figure , ax = plt.subplots(1,2, figsize = (16,8))

ax[0].imshow(img1, cmap = 'gray')
ax[1].imshow(img2, cmap = 'gray')
```

```python
#SIFT
sift = cv.SIFT_create()

kp1, descriptor1 = sift.detectAndCompute(img1, None)
kp2, descriptor2 = sift.detectAndCompute(img2, None)

#Coincidencias de caracteristicas
bf = cv.BFMatcher(cv.NORM_L1, crossCheck = True)

match1 = bf.match(descriptor1,descriptor2)
match1 = sorted(match1, key = lambda x: x.distance)

img3 = cv.drawMatches(img1, kp1, img2, kp2, match1[:500], img2, flags = 2)
plt.figure(figsize=(9,9))
plt.imshow(img3)
```

#Conclusión


Este archivo implementa el algoritmo SIFT para detectar puntos de interés locales en una imagen, los cuales son invariantes ante cambios de escala y rotación. El código realiza el preprocesamiento en escala de grises, calcula descriptores únicos para cada característica y visualiza gráficamente los resultados finales mediante el uso de la librería OpenCV.

En conclusión, el notebook demuestra la efectividad de SIFT para identificar estructuras específicas mediante un proceso de cuatro etapas que garantiza la estabilidad de los atributos detectados. Esta técnica resulta fundamental en la visión computacional, ya que permite generar representaciones consistentes que facilitan el análisis de escenas y el reconocimiento de objetos en diversas perspectivas.
