# Algoritmo Sobel

## Fundamentos Matemáticos

El operador de Sobel es un filtro de detección de bordes que aproxima el gradiente de intensidad de una imagen mediante dos convoluciones: una horizontal y otra vertical.

* Kernels de Sobel
* Magnitud del Gradiente
* Dirección del Gradiente

## Métricas

* Magnitud del gradiente: intensidad del borde.
* Dirección del gradiente: orientación del borde.
* Umbralización (opcional): determina qué píxeles se consideran bordes.

## Descripción formal o narrativa

Sobel es un operador de procesamiento de imágenes que calcula derivadas discretas en las direcciones horizontal y vertical mediante convolución con máscaras de tamaño 3x3.

Su salida es una imagen donde los cambios bruscos de intensidad aparecen resaltados.

## Pseudocódigo



```python
Sobel(imagen)

Aplicar padding

Para cada pixel:

    gx = convolucion(region, Gx)

    gy = convolucion(region, Gy)

    magnitud = sqrt(gx² + gy²)

Normalizar resultado

Retornar imagen_bordes
```

## Algoritmo


```python
import numpy as np

def sobel_filter(image):
    # Kernels de Sobel
    Gx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    Gy = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])

    rows, cols = image.shape
    output = np.zeros((rows, cols))

    # Aplicar padding simple para evitar errores en bordes
    padded_img = np.pad(image, (1, 1), mode='constant')

    for i in range(rows):
        for j in range(cols):
            region = padded_img[i:i+3, j:j+3]
            gx = np.sum(region * Gx)
            gy = np.sum(region * Gy)
            output[i, j] = np.sqrt(gx**2 + gy**2)

    # Normalización a 0-255
    output = (output / output.max()) * 255
    return output.astype(np.uint8)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color

# 1. Cargar una imagen de muestra (ej. astronauta) y convertir a gris
image = color.rgb2gray(data.astronaut())

# 2. Aplicar la función sobel_filter que definimos anteriormente
# (Asegúrate de que la función sobel_filter esté definida en el mismo script)
bordes = sobel_filter(image)

# 3. Visualización comparativa
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Imagen Original")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Resultado: Operador de Sobel")
plt.imshow(bordes, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
