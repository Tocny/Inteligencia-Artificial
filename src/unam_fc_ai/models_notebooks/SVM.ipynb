# Algoritmo de SVM

## Fundamentos Matemáticos

* Hiperplano separador: frontera que divide las clases. w * x + b = 0
* Función de decisión: determina la clase de un nuevo punto.
* Margen: distancia entre el hiperplano y los vectores de soporte.
* Objetivo: maximizar el margen para lograr una mejor generalización.

## Métricas

* Margen: distancia entre el hiperplano y los vectores de soporte.
* Accuracy: porcentaje de clasificaciones correctas.
* Precision: de los positivos predichos, cuántos fueron correctos.
* Recall: de los positivos reales, cuántos detectó.
* F1-Score: balance entre Precision y Recall.

## Descripción formal o narrativa

SVM intenta dibujar una línea divisoria entre dos grupos de datos (hiperplano). No solo busca separarlos, sino hacerlo con la mayor distancia posible respecto a los puntos más cercanos de cada clase, llamados vectores de soporte.

## Pseudocódigo



```python
SVM(X, y)

Inicializar pesos w y sesgo b

Mientras no termine el entrenamiento:

    Para cada ejemplo:

        Si cumple el margen:
            Actualizar regularización

        Si no cumple:
            Actualizar pesos y sesgo

Retornar w y b
```

## Algoritmo


```python
import numpy as np
from sklearn.svm import SVC # La versión profesional

class SVM_Manual:
    def __init__(self, learning_rate=0.001, lambda_param=0.01, n_iters=1000):
        self.lr = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.w = None
        self.b = None

    def fit(self, X, y):
        # Convertir etiquetas a [-1, 1] si vienen como [0, 1]
        y_ = np.where(y <= 0, -1, 1)
        n_samples, n_features = X.shape
        self.w = np.zeros(n_features)
        self.b = 0

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                # Condición de la pérdida Hinge
                condition = y_[idx] * (np.dot(x_i, self.w) - self.b) >= 1
                if condition:
                    self.w -= self.lr * (2 * self.lambda_param * self.w)
                else:
                    self.w -= self.lr * (2 * self.lambda_param * self.w
                                         - np.dot(x_i, y_[idx]))
                    self.b -= self.lr * y_[idx]

    def predict(self, X):
        approx = np.dot(X, self.w) - self.b
        return np.sign(approx)
```

## Complejidad

Complejidad Temporal

* Entrenamiento: entre O(n²·d) y O(n³·d)
    * n = número de ejemplos.
    * d = número de características.
    * Puede ser costoso para datasets muy grandes.
* Predicción: O(v·d)
    * v = número de vectores de soporte.
    * Muy rápida una vez entrenado.

Complejidad Espacial

* Espacio: O(n²)
    * Debe almacenar información de similitud entre muestras.
