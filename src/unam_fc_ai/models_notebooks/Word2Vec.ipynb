# Algoritmo Word2Vec

## Fundamentos Matemáticos

Word2Vec aprende una representación vectorial (embedding) de cada palabra. La idea es que palabras que aparecen en contextos similares tendrán vectores parecidos.

* Función Objetivo: Maximizar la probabilidad de observar las palabras de contexto dada una palabra central
* Softmax: Calcula la probabilidad de cada palabra del vocabulario.

## Métricas

* Cross-Entropy Loss: mide qué tan bien predice el contexto.
* Similitud Coseno: mide cercanía entre embeddings.

CosSim(A,B)= (A * B) / ||A|| ||B||

## Descripción formal o narrativa

Word2Vec es una red neuronal de una capa oculta que aprende vectores densos para representar palabras. La arquitectura más común es Skip-Gram, donde una palabra central intenta predecir sus palabras vecinas.

Imagina que lees millones de textos.

Si constantemente aparecen frases como:

* Messi anota gol
* Cristiano anota gol
* Delantero anota gol

el modelo aprende que “Messi”, “Cristiano” y “Delantero” suelen aparecer en contextos parecidos.

Por eso sus vectores terminan cerca en el espacio de embeddings.

## Pseudocódigo



```python
Word2Vec_SkipGram(Corpus)

Inicializar W1 y W2

Para cada palabra central:

    Obtener palabras de contexto

    Forward:
        h = W1[palabra]
        y_pred = Softmax(h * W2)

    Calcular error

    Backpropagation

    Actualizar W1 y W2

Retornar embeddings
```

## Algoritmo


```python
import numpy as np

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

class Word2VecSimplified:
    def __init__(self, vocab_size, embed_dist):
        self.n = embed_dist # Dimensiones
        self.v = vocab_size # Tamaño vocabulario
        # Inicialización de pesos (He initialization)
        self.w1 = np.random.randn(self.v, self.n)
        self.w2 = np.random.randn(self.n, self.v)

    def forward(self, x):
        # x es un vector one-hot o el índice de la palabra
        h = self.w1[x]
        u = np.dot(h, self.w2)
        y_c = softmax(u)
        return y_c, h, u

    def backward(self, e, h, x):
        # Gradientes para actualizar los pesos
        dw2 = np.outer(h, e)
        dw1 = np.dot(self.w2, e.T)

        # Actualización simple (tasa de aprendizaje 0.01)
        self.w2 = self.w2 - (0.01 * dw2)
        self.w1[x] = self.w1[x] - (0.01 * dw1.T)

# Ejemplo de uso conceptual:
# 1. Definimos vocabulario: {'gol': 0, 'messi': 1, 'futbol': 2}
# 2. Si entra 'messi' (index 1) y el contexto es 'gol' (index 0)
modelo = Word2VecSimplified(vocab_size=3, embed_dist=2)
y_pred, h, u = modelo.forward(1)

# El error es (predicción - real) donde real es un one-hot del contexto
error = y_pred.copy()
error[0] -= 1

modelo.backward(error, h, 1)
print("Vector actualizado de 'Messi':", modelo.w1[1])
```

## Complejidad

Complejidad Temporal (Entrenamiento)

O(E · T · c · log V)

donde:

* E: épocas.
* T: palabras del corpus.
* c: tamaño de ventana.
* V: vocabulario.

Complejidad Espacial

O(V · D)

donde:

* V: vocabulario.
* D: dimensión del embedding.
