# Algoritmo de K-NN

## Fundamentos Matemáticos

* Distancia Euclidiana: mide qué tan cerca están dos puntos.
* Regla de decisión: se seleccionan los k vecinos más cercanos y se asigna la clase con más votos.
* Parámetro k: controla cuántos vecinos participan en la decisión.

## Métricas

* Accuracy: porcentaje de clasificaciones correctas.
* Precision: de los positivos predichos, cuántos fueron correctos.
* Recall: de los positivos reales, cuántos detectó.
* F1-Score: balance entre Precision y Recall.

## Descripción formal o narrativa

k-NN funciona como preguntar la opinión a los vecinos más cercanos. Para clasificar un nuevo punto, busca los k ejemplos más parecidos y asigna la clase que obtenga la mayoría de votos.

## Pseudocódigo



```python
KNN(Dataset, NuevoPunto, k)

Para cada dato del dataset:
    Calcular distancia al nuevo punto

Ordenar distancias

Seleccionar los k vecinos más cercanos

Contar las etiquetas de esos vecinos

Retornar la etiqueta con más votos
```

## Algoritmo


```python
import numpy as np
from collections import Counter

class KNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        # En k-NN el "entrenamiento" es simplemente almacenar los datos
        # Complejidad en tiempo: O(1)
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        # Predecimos para cada punto en el conjunto de entrada
        predictions = [self._predict(x) for x in X]
        return np.array(predictions)

    def _predict(self, x):
        # 1. Calcular distancias entre x y todos los puntos en X_train
        # Usamos la norma euclidiana: ||x1 - x2||
        distances = [np.sqrt(np.sum((x - x_train)**2)) for x_train in self.X_train]

        # 2. Obtener los índices de los k vecinos más cercanos
        # np.argsort nos da los índices que ordenarían el arreglo
        k_indices = np.argsort(distances)[:self.k]

        # 3. Extraer las etiquetas de esos k vecinos
        k_nearest_labels = [self.y_train[i] for i in k_indices]

        # 4. Votación mayoritaria
        most_common = Counter(k_nearest_labels).most_common(1)
        return most_common[0][0]
```

## Complejidad

Complejidad Temporal

* Entrenamiento: O(1)
    * No aprende un modelo; solo almacena los datos.
* Predicción: O(n·d)
    * n = número de ejemplos.
    * d = número de características.
    * Debe calcular la distancia contra todos los puntos.

Complejidad Espacial

* Espacio: O(n·d)
    * Debe guardar todo el dataset en memoria.
