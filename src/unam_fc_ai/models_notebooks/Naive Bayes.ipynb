# Naive Bayes

## Fundamentos Matemáticos

Naive Bayes es un algoritmo probabilístico basado en el Teorema de Bayes. Su objetivo es calcular la probabilidad de que una instancia pertenezca a una clase dada la información observada.

La fórmula principal es:

P(C|X) = (P(X|C) · P(C)) / P(X)

Donde:

* P(C|X): probabilidad posterior.
* P(X|C): verosimilitud.
* P(C): probabilidad previa (prior).
* P(X): evidencia.

Como P(X) es igual para todas las clases, normalmente se compara:

P(C|X) ∝ P(X|C) · P(C)

## Métricas

* Accuracy: porcentaje total de predicciones correctas.
    Accuracy = (TP + TN) / (TP + TN + FP + FN)
* Precision: de todo lo que el modelo predijo como positivo, cuánto era realmente positivo.
    Precision = TP / (TP + FP)
* Recall (Sensibilidad): de todos los positivos reales, cuántos logró detectar.
    Recall = TP / (TP + FN)
* F1-Score: balance entre Precision y Recall.
    F1 = 2 × (Precision × Recall) / (Precision + Recall)

## Descripción formal o narrativa

Naive Bayes funciona como un sistema de votación probabilística.

Primero aprende cómo son los datos de cada clase calculando:

* Probabilidad previa (prior).
* Promedios.
* Desviaciones estándar.

Cuando llega una nueva instancia, calcula qué tan probable es observar cada característica dentro de cada clase usando la distribución normal. Después combina todas esas probabilidades y asigna la clase con la probabilidad más alta.

## Pseudocódigo



```python
Entrenar_Naive_Bayes(X, y)

Para cada clase C:

    Obtener datos de la clase C

    Calcular Prior(C)

    Calcular Promedio(C)

    Calcular Desviación(C)

Guardar estadísticas

------------------------------------------------

Predecir(instancia)

Para cada clase C:

    log_prob = log(Prior(C))

    Para cada característica x_i:

        Calcular PDF_Gaussiana(x_i)

        log_prob += log(PDF_Gaussiana)

Retornar clase con mayor probabilidad
```

## Algoritmo


```python
def entrenar_naive_bayes(X, y):
    resumen = {}
    for clase in np.unique(y):
        datos_clase = X[y == clase]
        resumen[clase] = {
            'prior': len(datos_clase) / len(X),
            'promedios': datos_clase.mean(axis=0),
             'desviaciones': datos_clase.std(axis=0)
        }
    return resumen

def pdf_gaussiana(x, promedio, desviacion):
    # Función de densidad de probabilidad para variables continuas
    exponente = np.exp(-((x - promedio)**2 / (2 * desviacion**2)))
    return (1 / (np.sqrt(2 * np.pi) * desviacion)) * exponente

def predecir(resumen, instancia):
    probabilidades = {}
    for clase, stats in resumen.items():
        # Iniciamos con el log del Prior para evitar problemas de precisión
        log_prob = np.log(stats['prior'])
        for i in range(len(instancia)):
            p_x_dado_c = pdf_gaussiana(instancia[i], stats['promedios'][i], stats['desviaciones'][i])
            log_prob += np.log(p_x_dado_c)
        probabilidades[clase] = log_prob
    return max(probabilidades, key=probabilidades.get)
```

## Complejidad

Complejidad Temporal

* Entrenamiento: O(N × M)
    * N = número de instancias (filas del dataset).
    * M = número de características (columnas).
* Predicción: O(K × M)
    * K = número de clases.
    * M = número de características.

Complejidad Espacial

* Espacio: O(K × M)
    * K = número de clases.
    * M = número de características.
