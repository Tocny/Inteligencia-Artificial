# Algoritmo de Markov

## Fundamentos Matemáticos

* Propiedad de Markov: El siguiente estado depende únicamente del estado actual, no de los estados anteriores.
    P(Xt+1 | Xt)
* Matriz de Transición (P): Matriz donde cada elemento P(i,j) representa la probabilidad de pasar del estado i al estado j. La suma de cada fila debe ser igual a 1.
* Vector de Estado (π): Representa la probabilidad de encontrarse en cada estado en un instante dado.
    π(n+1) = π(n) · P
* Estado Estacionario: Distribución de probabilidades que permanece constante después de muchas transiciones.
    πP = π
* Idea principal: Las cadenas de Markov modelan sistemas donde el futuro depende únicamente del presente y utilizan probabilidades de transición para predecir la evolución del sistema.

## Métricas

* Probabilidad de transición: probabilidad de pasar de un estado a otro.
* Estado estacionario: distribución estable a la que converge la cadena después de muchas transiciones.
* Irreducibilidad: capacidad de llegar de cualquier estado a cualquier otro.
* Aperiodicidad: evita ciclos fijos y permite convergencia.

## Descripción formal o narrativa

Una Cadena de Markov funciona como un sistema de predicción basado en probabilidades. Cada estado tiene ciertas probabilidades de transición hacia otros estados. El siguiente estado se selecciona aleatoriamente usando esas probabilidades.

Ejemplo: si hoy está Soleado, existe un 80% de probabilidad de que mañana siga Soleado.

## Pseudocódigo



```python
Inicializar estados y probabilidades

EstadoActual = EstadoInicial

Mientras haya pasos:

    Obtener probabilidades del EstadoActual

    Seleccionar siguiente estado según probabilidades

    Actualizar EstadoActual

    Guardar resultado

Retornar historial
```

## Algoritmo


```python
import random

# Definición manual del modelo
transiciones = {
    "Soleado": [("Soleado", 0.8), ("Nublado", 0.15), ("Lluvioso", 0.05)],
    "Nublado": [("Soleado", 0.2), ("Nublado", 0.5), ("Lluvioso", 0.3)],
    "Lluvioso": [("Soleado", 0.2), ("Nublado", 0.2), ("Lluvioso", 0.6)]
}

def siguiente_estado(estado_actual):
    opciones = transiciones[estado_actual]
    r = random.random()  # número entre 0 y 1
    acumulado = 0

    for estado, prob in opciones:
        acumulado += prob
        if r <= acumulado:
            return estado

def simular_clima(estado_inicial, dias):
    estado = estado_inicial
    historial = [estado]

    for _ in range(dias):
        estado = siguiente_estado(estado)
        historial.append(estado)

    return historial


# Ejemplo
resultado = simular_clima("Soleado", 10)
print("Simulación del clima:")
print(resultado)
```

## Complejidad

Complejidad Temporal

* Predicción de un paso: O(S²)
    * S = número de estados.
    * Se utilizan operaciones sobre la matriz de transición.

Complejidad Espacial

* Espacio: O(S²)
    * S = número de estados.
    * Se debe almacenar la matriz completa de probabilidades de transición.
