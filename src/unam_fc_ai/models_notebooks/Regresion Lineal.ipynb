# Algoritmo de Regresión Lineal

## Fundamentos Matemáticos

* Hipótesis del modelo: relación lineal entre variables.
* Función de costo (MSE): mide el error promedio al cuadrado.
* Gradiente descendente: actualiza los parámetros para minimizar el error.

## Métricas

* MSE (Mean Squared Error): error cuadrático medio.
* RMSE: raíz del error cuadrático medio.
* MAE: error absoluto medio.
* R²: porcentaje de variabilidad explicada por el modelo.
* Error residual: diferencia entre valor real y predicho.

## Descripción formal o narrativa

La regresión lineal es un algoritmo supervisado que modela la relación entre una variable dependiente y una o más variables independientes mediante una función lineal.

## Pseudocódigo



```python
RegresionLineal(X, y)

Inicializar pesos w y sesgo b

Para cada iteración:

    Calcular predicciones

    Calcular error

    Calcular gradientes

    Actualizar w y b

Retornar modelo
```

## Algoritmo


```python
class RegresionLineal:
    def __init__(self, alpha=0.01, iteraciones=1000):
        self.alpha = alpha
        self.iteraciones = iteraciones
        self.w = None
        self.b = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.w = np.zeros(n_features)
        self.b = 0

        for _ in range(self.iteraciones):
            # Predicción lineal
            y_pred = np.dot(X, self.w) + self.b

            # Cálculo de gradientes
            dw = (1 / n_samples) * np.dot(X.T, (y_pred - y))
            db = (1 / n_samples) * np.sum(y_pred - y)

            # Actualización
            self.w -= self.alpha * dw
            self.b -= self.alpha * db

    def predict(self, X):
        return np.dot(X, self.w) + self.b
```

## Complejidad

Complejidad Temporal

* Entrenamiento: O(k·n·d)
    * k = iteraciones.
    * n = ejemplos.
    * d = características.
* Predicción: O(d)
    * Solo calcula el producto punto entre datos y pesos.

Complejidad Espacial

* Espacio: O(d)
    * Solo almacena pesos y sesgo del modelo.
