# Multiplicación de Matrices

## Fundamentos Matemáticos

La multiplicación de matrices consiste en combinar dos matrices A y B para obtener una nueva matriz C.

Cada elemento de la matriz resultado se calcula como: se multiplica la fila i de A por la columna j de B y se suman los resultados.


## Métricas

* Número de multiplicaciones: n^3.
* Número de sumas: aproximadamente n^3.
* Escalabilidad: el costo crece rápidamente cuando aumenta n.

## Descripción Formal

Dadas dos matrices cuadradas, se construye una matriz resultado donde cada posición (i,j) se obtiene mediante el producto punto entre la fila i de A y la columna j de B.

## Pseudocódigo


```python
MultiplicarMatrices(A, B)

n ← tamaño de la matriz

Crear matriz C de tamaño n×n

Para i = 1 hasta n:
    Para j = 1 hasta n:
        C[i,j] ← 0

        Para k = 1 hasta n:
            C[i,j] ← C[i,j] +
                      A[i,k] * B[k,j]

Retornar C
```

## Algoritmo

```python
import numpy as np
import time

# Multiplicación de Matrices de manera normal
def multiplicacion_tradicional(A, B):
    n = len(A)
    C = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C


def probar_multiplicacion(n):
    print(f"\nMatrices de tamaño {n}x{n}")
    # Generamos nuestras matrices de tamaño n con valores aleatorios
    A = np.random.randint(0, 100, (n, n))
    B = np.random.randint(0, 100, (n, n))
    inicio = time.time()
    C = multiplicacion_tradicional(A, B)
    fin = time.time()
    print("Tiempo de ejecución:", fin - inicio, "segundos")
    if n > 10:
        print("Resultados parciales para nuestro matriz (tamaño de 5x5):")
        print(C[:5, :5])
    else:
        print("Resultado completo de toda la matriz:")
        print(C)

# Nuestras pruebas pedidas para nuestro ejercicio y hacer nuestro respectivo
# análisis
probar_multiplicacion(10)
probar_multiplicacion(100)
```

## Complejidad

Complejidad Temporal: O(n^3)

Complejidad Espacial: O(n^2)
