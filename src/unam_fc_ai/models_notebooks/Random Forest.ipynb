# Algoritmo Random Forest

## Fundamentos Matemáticos

* Se basa en Bagging (Bootstrap Aggregating).
* Entrena múltiples árboles sobre muestras aleatorias del dataset.
* La predicción final se obtiene mediante voto mayoritario.

## Métricas

* Accuracy: porcentaje de aciertos.
* Precision: qué tan correctos son los positivos predichos.
* Recall: qué tantos positivos reales detecta.
* F1-Score: balance entre precisión y recall.
* OOB Error (Out-of-Bag): estimación interna del error usando muestras no seleccionadas.

## Descripción formal o narrativa

En lugar de confiar en un solo árbol de decisión, Random Forest crea muchos árboles diferentes. Cada árbol vota por una respuesta y la clase más votada se convierte en la predicción final.

## Pseudocódigo



```python
RandomForest(X, y)

Para cada árbol:

    Crear muestra Bootstrap

    Entrenar árbol

Para cada predicción:

    Obtener voto de cada árbol

    Elegir la clase más votada

Retornar predicción
```

## Algoritmo


```python
import numpy as np
from collections import Counter

class RandomForest:
    def __init__(self, n_trees=10, max_depth=5, n_features=None):
        self.n_trees = n_trees
        self.max_depth = max_depth
        self.n_features = n_features
        self.trees = []

    def fit(self, X, y):
        self.trees = []
        for _ in range(self.n_trees):
            # 1. Crear muestra Bootstrap
            n_samples = X.shape[0]
            indices = np.random.choice(n_samples, n_samples, replace=True)
            X_sample, y_sample = X[indices], y[indices]

            # 2. Entrenar árbol (usando tu clase previa)
            # Nota: Para un RF real, el árbol debería elegir características al azar en cada nodo.
            tree = ArbolDecision(max_depth=self.max_depth)
            tree.fit(X_sample, y_sample)
            self.trees.append(tree)

    def predict(self, X):
        # Obtener predicciones de cada árbol
        # tree_preds será una lista de arreglos: [[p1, p2], [p1, p2], ...]
        tree_preds = np.array([tree.predict(X) for tree in self.trees])

        # Transponer para tener las predicciones por cada punto: [[árbol1, árbol2], ...]
        tree_preds = np.swapaxes(tree_preds, 0, 1)

        # Votación mayoritaria
        predictions = [Counter(p).most_common(1)[0][0] for p in tree_preds]
        return np.array(predictions)
```

## Complejidad

Complejidad Temporal

* Entrenamiento: O(B · n · d · log(n))
    * B = número de árboles.
    * n = ejemplos.
    * d = características.
* Predicción: O(B · profundidad)
    * Se consulta cada árbol del bosque.

Complejidad Espacial

* Espacio: O(B · nodos)
    * Deben almacenarse todos los árboles entrenados.
